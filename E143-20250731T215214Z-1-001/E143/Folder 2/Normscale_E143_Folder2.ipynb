```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the file
file_path = r'C:\Users\castl\Downloads\Data_Lionel\Data_Lionel\E143_data\2...HEPData-ins393667-v1-csv\Table2.csv'
```

```python
#Read the data, skipping the header lines that start with #
data = pd.read_csv(file_path, comment='#')
```

```python
#Extract the data columns
X = data['X']
G1 = data['G1']
stat_pos = data['stat +']
stat_neg = data['stat -']
```

```python
#Create the plot
plt.figure(figsize=(10, 6))

#Plot the data points with error bars (symetric in data)
plt.errorbar(X, G1, yerr=stat_pos, fmt='bo', capsize=3, capthick=1.5, 
             markersize=6, linewidth=1.5, elinewidth=1.5)

#Style
plt.xlabel('X', fontsize=14, fontweight='bold')
plt.ylabel('g', fontsize=14, fontweight='bold')
plt.title('g vs x, Q=3 GeV,  assuming g/f is independent of Q\ne + p  e + X', 
          fontsize=14, fontweight='bold')

#Set the axis limits to contain the entire error bars
x_min, x_max = X.min(), X.max()
g1_with_errors_min = (G1 - stat_pos).min()
g1_with_errors_max = (G1 + stat_pos).max()

#Add some padding to the limits
x_padding = (x_max - x_min) * 0.05
g1_padding = (g1_with_errors_max - g1_with_errors_min) * 0.05

plt.xlim(x_min - x_padding, x_max + x_padding)
plt.ylim(g1_with_errors_min - g1_padding, g1_with_errors_max + g1_padding)

#Add grid
plt.grid(True, alpha=0.3)

#Add a horizontal line at y=0 for reference
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)

#Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
#Summary
print(f"Number of data points: {len(X)}")
print(f"X range: {X.min():.3f} to {X.max():.3f}")
print(f"G1 range: {G1.min():.3f} to {G1.max():.3f}")
print(f"Average statistical error: {stat_pos.mean():.3f}")
```
