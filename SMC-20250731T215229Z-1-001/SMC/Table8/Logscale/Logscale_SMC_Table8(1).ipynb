```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the CSV file
file_path=r'C:\Users\castl\Downloads\Data_Lionel\Data_Lionel\SMC_data\HEPData-ins471981-v1-csv\Table8.csv'
df=pd.read_csv(file_path, comment='#')
```

```python
#Calculate x as the midpoint of x_low and x_high
x=(df['X LOW']+df['X HIGH'])/2
```

```python
#Extract G1 values and statistical errors
g1=df['G1']
stat_error_plus=df['stat +']
stat_error_minus=df['stat -']
```

```python
#Error bar calculation
stat_error=(stat_error_plus+abs(stat_error_minus))/2
```

```python
# Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(x, g1, yerr=stat_error, fmt='o', capsize=5, capthick=2, 
             markersize=6, linewidth=2, label='G1(NS)')

# Set labels and title
plt.xlabel('x', fontsize=12)
plt.ylabel('G1(NS)', fontsize=12)
plt.title('Spin-dependent Structure Function G1(NS) at Q² = 10 GeV²', fontsize=14)

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Add legend
plt.legend()

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()

# Print some basic statistics
print("Data Summary:")
print(f"Number of data points: {len(x)}")
print(f"x range: {x.min():.3f} to {x.max():.3f}")
print(f"G₁(D) range: {g1.min():.3f} to {g1.max():.3f}")
```
