```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the CSV file
file_path=r'C:\Users\castl\Downloads\Data_Lionel\Data_Lionel\SMC_data\HEPData-ins471981-v1-csv\Table11.csv'
```

```python
# Read the data, skipping the header comments
df=pd.read_csv(file_path, comment='#')
```

```python
#Calculate x as the midpoint of x_low and x_high
x=(df['X LOW']+df['X HIGH'])/2
```

```python
# Get G1 values
g1=df['G1(C=NS)']
```

```python
# Get error values for error bars
stat_plus=df['stat +']
stat_minus=df['stat -']
```

```python
#Calculate total errors (combining statistical and systematic)
error_plus=np.sqrt(stat_plus**2)
error_minus=np.sqrt(stat_minus**2)
```

```python
#Plot and Data
#Create the plot
plt.figure(figsize=(10, 6))
plt.errorbar(x, g1, yerr=[error_minus, error_plus], fmt='o', capsize=5, 
             capthick=2, markersize=6, linewidth=2, label='G1(NS)')

plt.xlabel('x', fontsize=12)
plt.ylabel('G1(C=NS)', fontsize=12)
plt.title('Nonsinglet Spin-Dependent Structure Function G1(NS) vs x\n(Q² = 10 GeV²)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()

# Adjust layout and show plot
plt.tight_layout()
plt.show()

# Print some basic statistics
print("Data Summary:")
print(f"Number of data points: {len(x)}")
print(f"x range: {x.min():.4f} to {x.max():.4f}")
print(f"G1 range: {g1.min():.4f} to {g1.max():.4f}")
```
